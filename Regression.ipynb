```python
# Q1. From the cars dataset, predict the estimated distance of a car when the speed is given
```

```python
from pydataset import data

car = data('cars')
car.head()
```

```python
car.shape
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
```

```python
# Collecting x and y valyes

x = cars['speed'].values
y = cars['dist'].values

print(x)
print(y)
```

```python
type(x)
```

```python
plt.scatter(x,y)
plt
```

```python
cars.corr()
```

```python
# Since speed and dist correlation is 0.8 which is towards 1, it means strongly correlated. Therefore we can go for a model

n = len(x)
x = x.reshape(n,1)

# Model initialization

reg = LinearRegression()

# Data Fitting

reg = reg.fit(x,y)
```

```python
# Find the slope

slope = reg.coef_
slope
```

```python
# Find the intercept

c = reg.intercept_
c
```

```python
# Thus the equation will be y = 3.932 * speed - 17.5790
```

```python
# Pridict y

y_predict = reg.predict(x)
plt.scatter(x,y)
plt.plot(x,y_predict)
```

```python
# Model Evaluation

rmse = np.sqrt(mean_squared_error(y,y_predict))
r2 = reg.score(x,y)

print("RMSE = ",rmse)
print("R2 square = ",r2)
```

```python
# If root mean square value is 0 it means there is no error in the model therefore RMSE should be towards 0
# Here R square value neans, by this model only 65% priction value can be true and 35% value can not be and this may be because
# of the type of the road conditions, type of driver etc.
```

```python
# Using the model, pridict the distance 

user_input = [[int(input("Enter the Speed of the vehicle"))]]
print("The estimated distance is : ",reg.predict(user_input))
```
